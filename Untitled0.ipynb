# Reporte: Novedades del App Store Connect API – WWDC25

**Sesión:** Automate your development process with the App Store Connect API  

---

## ¿De qué trató esta sesión?

Durante la sesión, Apple presentó mejoras importantes al App Store Connect API. El enfoque principal fue facilitar la automatización de procesos clave en el desarrollo de aplicaciones, permitiendo a los equipos enfocarse más en crear experiencias de calidad para el usuario, y menos en tareas repetitivas.

---

## Novedades más relevantes

### 1. Webhooks: Tu app ahora puede reaccionar automáticamente

Hasta ahora era necesario consultar manualmente si había algún cambio relevante, como por ejemplo, si una build ya fue procesada. Con la introducción de los Webhooks, App Store Connect puede notificar automáticamente cuando ocurre un evento importante.

Ejemplos de eventos disponibles:
- Procesamiento completado de una build
- Aprobación de una versión en TestFlight
- Recepción de feedback por parte de testers

Esto permite conectar estos eventos con sistemas existentes, como flujos CI/CD o sistemas de alertas.  
**Nota:** Las notificaciones están firmadas con un secreto configurable, lo que garantiza su autenticidad.

---

### 2. Build Upload API: Subida de builds desde el pipeline

Apple lanzó una nueva API que permite subir builds directamente desde sistemas externos, sin depender de Xcode o Transporter. Esto abre la puerta a una mayor flexibilidad, ya que:

- Se puede usar cualquier lenguaje (Python, Node.js, etc.)
- Es compatible con cualquier flujo CI/CD
- Ofrece respuestas estructuradas y claras ante errores

El proceso consiste en:
1. Crear el objeto `BuildUpload`
2. Registrar el archivo (nombre, tamaño, etc.)
3. Enviar el binario a la URL proporcionada por Apple
4. Confirmar la carga para iniciar el procesamiento

Combinado con los Webhooks, este proceso permite automatizar el ciclo completo de entrega.

---

### 3. Feedback API: Acceso al feedback de TestFlight

Hasta ahora, el feedback de TestFlight debía revisarse manualmente. Con la nueva Feedback API, ahora es posible:

- Recibir retroalimentación en tiempo real mediante Webhooks (comentarios, errores, capturas de pantalla)
- Consultar detalles técnicos por API (modelo del dispositivo, sistema operativo, logs)
- Descargar archivos asociados como logs de fallos o capturas de pantalla

Esto permite a los equipos reaccionar de forma más rápida y comprender mejor la experiencia del usuario.

## Otros puntos relevantes

- Los Webhooks pueden registrarse desde la interfaz gráfica o mediante API, lo cual es útil para gestionar múltiples aplicaciones.
- También se presentaron mejoras para manejar recursos hospedados por Apple y cambios en los estados de versiones en el App Store.
- Todo el ciclo de vida de una app (subida, pruebas, feedback y publicación) ahora es más automatizable que nunca.

---

## Considerar

- Integrar la nueva Build Upload API en nuestros flujos de entrega continua.
- Automatizar el análisis del feedback de TestFlight para detectar y resolver problemas con mayor agilidad.
